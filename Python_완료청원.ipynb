```python
result = pd.DataFrame({"title":title, 'start':start, 'end':end, 'num':num, 'answer_date':ans})
result
```

```python
result['start'] = pd.to_datetime(result.start.map(lambda x: x.replace(". ","-")))
result['end'] = pd.to_datetime(result.end.map(lambda x: x.replace(". ","-")))
result['answer_date'] = pd.to_datetime(result.answer_date.map(lambda x: x.replace(". ","-")))
```
